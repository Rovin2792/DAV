```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('examples/mtcars.csv')
pd.set_option('display.max_columns', 100)  
pd.set_option('display.max_rows',None)     
```

```python
df
```

1. Extract the 'mpg' column from the dataframe & Plot the mpg values as a line plot

```python
mpg = df['mpg']
mpg.plot()
```

2. Select only the 'cyl', 'wt', and 'mpg' columns from the dataframe & Plot the selected dataframe columns

```python
df = df[['cyl', 'wt', 'mpg']]
df.plot()
```

3. Bar Plot- the count of cars by the number of cylinders

```python
cyl_counts = df['cyl'].value_counts()
cyl_counts.plot(kind='bar', color='green')
plt.title('Count of Cars by Cylinders')
plt.xlabel('Cylinders')
plt.ylabel('Count')
plt.show()
```

4. Scatter plot of weight vs mpg

```python
df.plot(kind="scatter",x="wt",y="mpg",figsize=(9,9),color="black",)
```

5. Box plot of MPG for each cylinder category

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='cyl', y='mpg', data=df)
plt.title('MPG by Cylinder Category')
plt.xlabel('Number of Cylinders')
plt.ylabel('MPG')
plt.show()
```

6. Histogram of MPG for all cars

```python
x = df['mpg']
plt.hist(x)
plt.title("Distribution of MPG in Cars Dataset")
plt.ylabel('Frequency')
plt.xlabel('MPG')
plt.show()
```

7. This pie chart shows the proportion of cars with 4, 6, or 8 cylinders.

```python
cyl_counts = df['cyl'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(cyl_counts,labels=cyl_counts.index,autopct='%1.1f%%',colors=['skyblue', 'lightgreen', 'orange'],
    startangle=140,explode=[0.1, 0, 0])
plt.title('Proportion of Cars by Number of Cylinders')
plt.show()

```

8. Pie Chart for Transmission Types

```python
df=pd.read_csv('examples/mtcars.csv')
```

```python
am_counts = df['am'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(am_counts,labels=['Automatic', 'Manual'],autopct='%1.1f%%',colors=['gold', 'lightcoral'],startangle=90)
plt.title('Proportion of Cars by Transmission Type')
plt.show()
```

9. Nested Pie Chart which needs Data preparation

```python
outer_labels = df['cyl'].value_counts().index
outer_sizes = df['cyl'].value_counts().values
inner_labels = df['gear'].value_counts().index
inner_sizes = df['gear'].value_counts().values
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(outer_sizes, labels=outer_labels, radius=1, autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'orange'])
ax.pie(inner_sizes, labels=inner_labels, radius=0.7, autopct='%1.1f%%', colors=['pink', 'purple', 'cyan'])
ax.set(aspect="equal")
plt.title('Nested Pie Chart: Cylinders (Outer) and Gears (Inner)')
plt.show()
```
