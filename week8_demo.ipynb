```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('examples/mtcars.csv')
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', None)
```

```python
df
```

```python
tab1=pd.crosstab(df['cyl'],df['gear'])
print("Tabl summarasing number of cars(Cylinders vs Gears)")
print(tab1)
```

```python
sns.heatmap(tab1, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Crosstab Heatmap: Cylinders vs Gears')
plt.xlabel('Gears')
plt.ylabel('Cylinders')
plt.gca().collections[0].colorbar.set_label('frequency')
```

A heatmap uses colors to show data patterns, making it easier to see trends, while a table shows exact numbers but can be harder to read quickly.

```python
tab2=df[['model','mpg']].sort_values('mpg',ascending=False)
print("Table displaying Car models(best to worst fuel effciency)")
print(tab2)
```

```python
tab3=df.groupby('cyl')['mpg'].mean()
print("Table displayng number of cylinders and average mpg")
print(tab3)
```

```python
sns.barplot(x ='cyl', y ='mpg', data = df,errorbar=None)
```

This means that as the number of cylinders increases, fuel efficiency (mpg) decreases.

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x='hp', y='mpg', data=df)
plt.title("MPG vs HP")
plt.show()
```

As horsepower increases, fuel efficiency (mpg) decreases.

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['mpg'], bins=10, kde=True)
plt.title("Distribution of MPG")
plt.show()
```

 Most vehicles have an mpg between 15 and 25, with fewer vehicles at the higher and lower ends of the range.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.histplot(df['mpg'], bins=20, kde=True, ax=axes[0], color='blue')
axes[0].set_title('MPG Distribution')
axes[0].set_xlabel('Miles Per Gallon (MPG)')
axes[0].set_ylabel('Count')

sns.boxplot(x='cyl', y='mpg', data=df, ax=axes[1], hue='cyl', palette='viridis', legend=False)
axes[1].set_title('MPG by Number of Cylinders')
axes[1].set_xlabel('Number of Cylinders')
axes[1].set_ylabel('Miles Per Gallon (MPG)')

plt.tight_layout()
plt.show()
```

The left plot shows the distribution of miles per gallon (mpg) across different vehicles, with most vehicles having an mpg between 15 and 25. The right plot compares mpg across different cylinder, showing that vehicles with fewer cylinders generally have higher fuel efficiency.

```python
sns.pairplot(df[['mpg', 'hp', 'wt', 'qsec']])
plt.show()
```

This pair plot shows relationships between miles per gallon (mpg), horsepower (hp), weight (wt), and quarter-mile time (qsec). We can see negative correlations between mpg and both hp and wt, meaning more powerful and heavier cars tend to have lower fuel efficiency.

```python
plt.figure(figsize=(6, 4))
sns.violinplot(x='cyl', y='mpg', data=df)
plt.title("MPG Distribution by Cylinder Count")
plt.show()
```

As the number of cylinders increases, fuel efficiency (mpg) decreases. Cars with 4 cylinders have the highest mpg spread, while 8-cylinder cars tend to have lower and more concentrated mpg values.

```python
plt.figure(figsize=(6, 4))
sns.regplot(x='wt', y='mpg', data=df)
plt.title("MPG vs Weight (Regression)")
plt.show()
```

As vehicle weight increases, fuel efficiency (mpg) decreases. The regression line confirms a negative correlation between weight and mpg.

```python
plt.figure(figsize=(6, 4))
sns.swarmplot(x='cyl', y='mpg', data=df)
plt.title("MPG by Cylinder Count (Swarm Plot)")
plt.show()
```

Fuel efficiency (mpg) clusters based on cylinder count. Cars with 4 cylinders achieve the highest mpg, while 8-cylinder vehicles are the least efficient.
