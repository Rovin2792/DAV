1. Identify top-selling products and their categories.Which products contribute most to revenue? Are there seasonal sales trends?

```python
top_selling = neworder.groupby(['ProductName', 'CategoryName'])['Quantity'].sum().reset_index()
top_selling = top_selling.sort_values(by='Quantity', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x='Quantity', y='ProductName', data=top_selling.head(10), hue='ProductName', dodge=False, palette='viridis', legend=False)
plt.xlabel('Total Quantity Sold')
plt.ylabel('Product Name')
plt.title('Top 10 Best-Selling Products')
plt.show()
```

Top selling products by quantity are Camembert Pierrot and Raclette Courdavault.

```python
neworder['Revenue'] = neworder['UnitPrice_x'] * neworder['Quantity'] * (1 - neworder['Discount'])
top_revenue = neworder.groupby('ProductName')['Revenue'].sum().reset_index()
top_revenue = top_revenue.sort_values(by='Revenue', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x='Revenue', y='ProductName', data=top_revenue.head(10), hue='ProductName', dodge=False, palette='coolwarm', legend=False)
plt.xlabel('Total Revenue')
plt.ylabel('Product Name')
plt.title('Top 10 Products by Revenue')
plt.show()
```

Top selling product by revenue is Cote de Blaye, which is almost double the next, Thuringer Rostbratwurst.

```python
category_sales = neworder.groupby('CategoryName')['Revenue'].sum().reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(x='Revenue', y='CategoryName', data=category_sales, palette='magma', hue='CategoryName', legend=False)
plt.xlabel('Total Revenue')
plt.ylabel('Category Name')
plt.title('Revenue by Product Category')
plt.show()
```

Total revenue is mainly generated by the categories of Beverages and Dairy Products.

2. Analyze customer orders by region and frequency.Identify and visualize high-value customers and regions

```python
order_frequency = neworder.groupby('ShipRegion')['OrderID'].count().reset_index().dropna()
order_frequency = order_frequency.sort_values(by='OrderID', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x='OrderID', y='ShipRegion',hue='ShipRegion',legend=False, data=order_frequency, order=order_frequency['ShipRegion'], palette='viridis')
plt.xlabel("Number of Orders")
plt.ylabel("Region")
plt.title("Order Frequency by Region")
plt.show()
```

The most number of orders come from the regions SP and ID.

```python
region_spending = neworder.dropna(subset=['ShipRegion']).groupby('ShipRegion')['TotalPrice'].sum().reset_index()
region_spending = region_spending.sort_values(by='TotalPrice', ascending=False)
region_spending_matrix = region_spending.set_index('ShipRegion')

plt.figure(figsize=(12, 8))
sns.heatmap(region_spending_matrix, cmap='coolwarm', annot=True, fmt=".2f", linewidths=0.5, cbar_kws={'label': 'Total Spending ($)'}, vmin=0)
plt.title("Total Spending by Region (Heatmap)")
plt.xlabel("Total Spending ($)")
plt.ylabel("Region")
plt.show()
```

From the heatmap, it is clear that region ID is where the most spending happens, followed by SP, which corresponds to their number of orders.

```python
neworder['TotalPrice'] = neworder['Quantity'] * neworder['UnitPrice_x'] * (1 - neworder['Discount'])
customer_revenue = neworder.groupby('CustomerID')['TotalPrice'].sum().reset_index()
customer_revenue = customer_revenue.sort_values(by='TotalPrice', ascending=False)
top_revenue_customers = customer_revenue.head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='TotalPrice', y='CustomerID', hue='CustomerID', legend=False, data=top_revenue_customers, palette='viridis')
plt.xlabel("Total Revenue ($)")
plt.ylabel("Customer ID")
plt.title("Top 10 Customers by Total Revenue")
plt.show()
```

Customers like QUICK, ERNSH, and SAVEA are the ones that generate the most revenue.

```python
customer_order_frequency = neworder.groupby('CustomerID')['OrderID'].count().reset_index()
customer_order_frequency = customer_order_frequency.sort_values(by='OrderID', ascending=False)
top_ordering_customers = customer_order_frequency.head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='OrderID', y='CustomerID',hue='CustomerID', legend=False, data=top_ordering_customers, palette='viridis')
plt.xlabel("Number of Orders")
plt.ylabel("Customer ID")
plt.title("Top 10 Customers by Order Frequency")
plt.show()
```

The number of orders and revenue generated from the old bar graph are related, as the top orders are from SAVEA, ERNSH, and QUICK.

```python
customer_spending_freq = customer_order_frequency.merge(neworder.groupby('CustomerID')['TotalPrice'].sum().reset_index(),on='CustomerID')

plt.figure(figsize=(10, 6))
sns.scatterplot(x='OrderID', y='TotalPrice', data=customer_spending_freq, alpha=0.7, color='blue')
sns.regplot(x='OrderID', y='TotalPrice', data=customer_spending_freq, scatter=False, color='red', line_kws={'linewidth':2})
plt.xlabel("Number of Orders")
plt.ylabel("Total Spending ($)")
plt.title("Correlation Between Order Frequency and Spending")
plt.show()
```

There is a positive correlation between the number of orders and total spending, with very few outliers.

3. Who are the employees with the highest sales or lowest sales? Which employees contribute to the revenue the most?

```python
employee_sales = neworder.groupby('EmployeeID')['TotalPrice'].sum().reset_index()
employee_sales = employee_sales.sort_values(by='TotalPrice', ascending=False)
top_employees = employee_sales.head(10)
bottom_employees = employee_sales.tail(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='EmployeeID', y='TotalPrice', data=top_employees, palette='Blues_r', hue='EmployeeID', legend=False)
plt.xlabel("Employee ID")
plt.ylabel("Total Sales ($)")
plt.title("Employees by Sales")
plt.show()
```

Employees 4, 3, and 1 are responsible for the highest total sales, whereas employees 5, 6, and 9 contribute very little.

```python
total_revenue = employee_sales['TotalPrice'].sum()
employee_sales['RevenueContribution (%)'] = (employee_sales['TotalPrice'] / total_revenue) * 100
employee_sales = employee_sales.sort_values(by='RevenueContribution (%)', ascending=False)
top_contributors = employee_sales.head(10)

plt.figure(figsize=(8, 8))
plt.pie(top_contributors['RevenueContribution (%)'], labels=top_contributors['EmployeeID'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('coolwarm', len(top_contributors)))
plt.title("Top 10 Employees Contribution to Total Revenue")
plt.axis('equal')
plt.show()
```

From the pie chart, it is easily visible that employees 4, 3, and 1 are the ones contributing almost 50% to the total revenue, whereas employees 5, 6, and 9 contribute less than 25%.
