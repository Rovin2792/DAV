```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("examples/sentiment.csv", index_col=0)
df
```

1. In the sentiment feature and text, trim white spaces and convert to lowercase.

```python
df['Sentiment']=df['Sentiment'].str.strip().str.lower()
df['Text']=df['Text'].str.strip().str.lower()
df
```

2. What are the sentiment categories? Tabulate and visualize the number of texts for each category.

```python
df['Sentiment'].unique()
```

```python
df['Sentiment'].nunique()
```

```python
tab = df.groupby('Sentiment')['Text'].count().sort_values(ascending = False).head(10)
plt.figure(figsize=(15,5))
sns.barplot(x = tab.index, y= tab.values)
```

The bar graph shows the number of texts for each sentiment category, with positive and joy having the highest counts and despair having the lowest.

3. Tabulate and Visualize the text length for each sentiment category.

```python
df['Text_length'] = df['Text'].apply(len)
text = df.groupby('Sentiment')['Text_length'].mean().sort_values(ascending = False).head(10)
plt.figure(figsize = (20,6))
sns.barplot(x=text.index, y= text.values)
```

The bar graph shows the number of texts for each sentiment category, with "winter magic" having the highest count and other categories having slightly lower but similar counts.

4. Find and display the text(s) with the maximum number of likes.

```python
tab = df.groupby('Text')['Likes'].sum().sort_values(ascending = False).head(10)
plt.figure(figsize = (10,6))
sns.barplot(y= tab.index, x = tab.values)
```

The bar graph shows the texts with the highest number of likes, where the top text has the most likes, and the rest follow in descending order

5. Find and display the text(s) with the maximum reposts. what is its sentiment?

```python
tab = df.groupby('Text')['Retweets'].sum().sort_values(ascending = False).head(10)
plt.figure(figsize = (10,6))
sns.barplot(y= tab.index, x = tab.values)
```

The bar graph shows the top 10 texts with the highest retweets, ranked in descending order.

6. Tabulate and Visualize the top 5 countries that have posted the most texts with a positive sentiment.

```python
tab = df[df['Sentiment']=='positive'].groupby('Country')['Sentiment'].count().sort_values(ascending = False).head(5)
sns.barplot(x = tab.index, y = tab.values)
```

The bar graph displays the top 5 countries that posted the most texts with a positive sentiment, with India and the USA having the highest counts.

10. Tabulate and visualize the years and countrywide tweets of the sentiments. Pick tweets with sentiments, â€˜Positive,â€™ â€˜Negative,â€™ and â€˜Neutralâ€™ only.

```python
year_country_data = df[['Year','Country','Sentiment']][(df['Sentiment'] == 'positive') |(df['Sentiment'] == 'negative') |(df['Sentiment'] == 'neutral') ]
print(year_country_data['Year'].unique())
year_country_data['Country'] = year_country_data['Country'].str.strip()
sns.barplot(x=year_country_data['Sentiment'].value_counts().index,y=year_country_data['Sentiment'].value_counts().values)
plt.title('Sentiment Distribution in 2023')
plt.show()
print('Sentiment distribution for 2023')
year_country_data['Sentiment'].value_counts()
fig, ax = plt.subplots(figsize=(10, 6))
country_data = year_country_data.drop('Year',axis=1).groupby('Country')['Sentiment'].value_counts().plot(kind='bar',ax=ax)
plt.title('Country wise Sentiment Distribution')
plt.ylabel('Count')
plt.xlabel('Country')
plt.show()
```

1. Sentiment Distribution in 2023: Most tweets were positive, followed by neutral and negative.
2. Country-wise Sentiment Distribution: The USA had the highest number of positive tweets, with other countries showing a mix of sentiments.
